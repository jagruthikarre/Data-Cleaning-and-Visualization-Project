```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv('2021_Cars_Aggregated.csv')

df.columns = df.columns.str.strip()

print("--- Dataset Overview ---")
print(f"Total rows: {len(df)}")
print(df[['Manufacturer', 'Fuel Type', 'Number of vehicles']].head())

df_clean = df.dropna(subset=['OBFCM Fuel consumption (l/100 km)', 'WLTP Fuel consumption (l/100 km)']).copy()

fuel_type_gap = df_clean.groupby('Fuel Type').agg(
    Avg_OBFCM_Fuel=('OBFCM Fuel consumption (l/100 km)', 'mean'),
    Avg_WLTP_Fuel=('WLTP Fuel consumption (l/100 km)', 'mean'),
    Avg_Pct_Gap=('percentage gap Fuel consumption (%)', 'mean'),
    Total_Vehicles=('Number of vehicles', 'sum')
).reset_index().sort_values(by='Avg_Pct_Gap', ascending=False)

print("\n--- Average Fuel Consumption Gap by Fuel Type ---")
print(fuel_type_gap.to_string(index=False))


significant_manufacturers = df_clean[df_clean['Number of vehicles'] > 100]
top_gaps_mfg = significant_manufacturers.groupby('Manufacturer')['percentage gap Fuel consumption (%)'].mean().reset_index()
top_gaps_mfg = top_gaps_mfg.sort_values(by='percentage gap Fuel consumption (%)', ascending=False).head(10)

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

fuel_type_melted = fuel_type_gap.melt(
    id_vars='Fuel Type',
    value_vars=['Avg_OBFCM_Fuel', 'Avg_WLTP_Fuel'],
    var_name='Test Type',
    value_name='Fuel Consumption (l/100 km)'
)
fuel_type_melted['Test Type'] = fuel_type_melted['Test Type'].map({'Avg_OBFCM_Fuel': 'Real-World (OBFCM)', 'Avg_WLTP_Fuel': 'Official Test (WLTP)'})

sns.barplot(
    data=fuel_type_melted,
    x='Fuel Consumption (l/100 km)',
    y='Fuel Type',
    hue='Test Type',
    ax=axes[0],
    palette='muted'
)
axes[0].set_title('Real-World (OBFCM) vs. Official (WLTP) Fuel Consumption', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Average Fuel Consumption (l/100 km)')
axes[0].set_ylabel('Fuel Type')

sns.barplot(
    data=top_gaps_mfg,
    x='percentage gap Fuel consumption (%)',
    y='Manufacturer',
    ax=axes[1],
    palette='Reds_r',
    hue='Manufacturer',
    legend=False
)
axes[1].set_title('Top 10 Manufacturers with Largest Fuel Consumption Gap\n(Min. 100 vehicles evaluated)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Average Percentage Gap (%)')
axes[1].set_ylabel('Manufacturer')

plt.tight_layout()

plt.savefig('fuel_efficiency_gap_report.png', dpi=300)
print("\n[Success] Analytics visualization saved as 'fuel_efficiency_gap_report.png'")

plt.show()
```
