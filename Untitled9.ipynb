```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv('CountriesSD.csv')

if df.columns[0].startswith('Unnamed') or df.columns[0] == '':
    df = df.iloc[:, 1:]

df.columns = df.columns.str.strip()

print("--- Dataset Summary ---")
print(f"Total records: {len(df)}")
print(df.info())

df_clean = df.dropna(subset=['Population', 'GDP per Capita']).copy()

df_clean['Total GDP'] = df_clean['Population'] * df_clean['GDP per Capita']

top_gdp_countries = df_clean.sort_values(by='Total GDP', ascending=False).head(10)

print("\n--- Top 5 Countries by Total GDP ---")
print(top_gdp_countries[['Country', 'Code', 'Population', 'GDP per Capita', 'Total GDP']].head(5).to_string(index=False))
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(
    data=top_gdp_countries,
    x='Total GDP',
    y='Country',
    ax=axes[0],
    palette='Blues_r',
    hue='Country',
    legend=False
)
axes[0].set_title('Top 10 Countries by Total Economic Output (Total GDP)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Total GDP (in USD)', fontsize=12)
axes[0].set_ylabel('Country', fontsize=12)

sns.scatterplot(
    data=df_clean,
    x='Population',
    y='GDP per Capita',
    ax=axes[1],
    alpha=0.7,
    color='teal',
    edgecolor='w'
)
axes[1].set_xscale('log')
axes[1].set_title('Population vs. GDP per Capita (Log Scale)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Population (Log Scale)', fontsize=12)
axes[1].set_ylabel('GDP per Capita (USD)', fontsize=12)

plt.tight_layout()

plt.savefig('global_economic_report.png', dpi=300)
print("\n[Success] Visualization report saved as 'global_economic_report.png'")

plt.show()
```
